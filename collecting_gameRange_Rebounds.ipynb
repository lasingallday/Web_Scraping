```python
from nba_py import game, team
import pandas as pd
import numpy as np
```

```python
# Constants
game_id = ['0021600849','0021600836']
tID = '1610612738'
```

```python
bs = game.Boxscore(game_id=game_id[0], season='2016-17', season_type='Regular Season', range_type='0', start_period='0', end_period='0', start_range='0', end_range='0')
df_total = bs.player_stats()
df_total = df_total[df_total['TEAM_CITY']=='Boston'.encode('utf-8')]
df_total.fillna(value=0)
```

```python
bs = game.Boxscore(game_id=game_id[1], season='2016-17', season_type='Regular Season', range_type='0', start_period='0', end_period='0', start_range='0', end_range='0')
df = bs.player_stats()
df = df[df['TEAM_CITY']=='Boston'.encode('utf-8')]
df.fillna(value=0)
```

```python
oreb_gameRange = pd.concat([df_total,df]).groupby(['PLAYER_NAME'], as_index=False)['OREB'].sum()
```
